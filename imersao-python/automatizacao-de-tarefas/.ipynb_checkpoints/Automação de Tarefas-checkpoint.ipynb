## 1: Buscar dados das ações automaticamente

```python
import yfinance as yf

ticker = input("Digite o código da ação: ")
print(ticker)
# dados = yf.Ticker("PETR4.SA").history("6mo")
dados = yf.Ticker(ticker).history("6mo")
fechamento = round(dados.Close, 2)
fechamento.plot()
fechamento
```

## 2: Realizar as análises solicitadas pelo gestor
- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2) # índice negativo

print(maxima)
print(minima)
print(atual)
```

## 3: Enviar automaticamente o e-mail com o resultado da análise
- abrir uma nova aba (Ctrl + T)
- entrar no Gmail
- clicar no botão Escrever
- definir o destinatário (Para) + Tab
- digitar o Assunto + Tab
- preencher o corpo e-mail
- enviar o e-mail

```python
pip install pyautogui
```

```python
import pyautogui
import pyperclip
```

```python
#dar uma pausa de 2 segundos entre os passos
pyautogui.PAUSE = 1.5

#abrir nova aba (ctrl+t)
pyautogui.hotkey("ctrl", "t")
    
#entrar no Gmail
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#abrir o Gmail
pyautogui.click(x=1443, y=134)
# pyperclip.copy("usuario")
# pyautogui.hotkey("ctrl", "v")
# pyautogui.click(x=1149, y=777)
# pyperclip.copy("senha")
# pyautogui.hotkey("ctrl", "v")
# pyautogui.click(x=1143, y=748)

#clicar no botão Escrever
pyautogui.click(x=127, y=225)

#definir Para (tab)
pyperclip.copy("email@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#digitar o Assunto (tab)
pyperclip.copy("Análise das Ações")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#preencher o corpo do e-mail
mensagem = f"""
Prezado gestor,

Seguem abaixo as análises das ações {ticker} dos últimos 6 meses, conforme solicitado:

Cotação máxima: R$ {maxima}
Cotação mínima: R$ {minima}
Cotação atual:  R$ {atual}

Qualquer dúvida, fico à disposição!

Atenciosamente,
Leandro.
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

#enviar e-mail
pyautogui.click(x=1180, y=970)
print("E-mail enviado com sucesso!")
```
